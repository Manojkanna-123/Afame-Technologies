## Data Visualization

### Distribution of Employees Across Various Departments

```python
# Department-wise Employee Count
department_count = data['Department'].value_counts()

# Plotting the bar chart
plt.figure(figsize=(8, 5.5))
department_count.plot(kind='bar', color='skyblue')
plt.title('Department-wise Employee Count')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Age Analysis of Employees

```python
# Plotting the histogram for age distribution
plt.figure(figsize=(8, 5.5))
plt.hist(data['Age'], bins=15, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Analyzing Attrition Trends Across Job Satisfaction Levels

```python
# Stacked bar chart visualizes the relationship between job satisfaction levels and attrition
job_satisfaction_vs_attrition = data.groupby(['JobSatisfaction', 'Attrition']).size().unstack()
total_counts = job_satisfaction_vs_attrition.sum(axis=1)
job_satisfaction_vs_attrition_percentage = (job_satisfaction_vs_attrition.T / total_counts).T * 100
plt.figure(figsize=(8,5.5))
job_satisfaction_vs_attrition_percentage.plot(kind='bar', stacked=True, colormap='viridis', width=0.7, edgecolor='black')
plt.title('Job Satisfaction vs. Attrition', fontsize=16)
plt.xlabel('Job Satisfaction Level', fontsize=12)
plt.ylabel('Percentage of Employees (%)', fontsize=12)
plt.legend(title='Attrition', fontsize=10)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### showing Average Monthly Income Across Job Roles

```python
# Horizontal bar chart displays the average monthly income for each job role
income_by_job_role = data.groupby('JobRole')['MonthlyIncome'].mean().sort_values()
plt.figure(figsize=(8, 5.5))
income_by_job_role.plot(kind='barh', color='skyblue', edgecolor='black')
plt.title('Average Monthly Income by Job Role', fontsize=16)
plt.xlabel('Average Monthly Income', fontsize=12)
plt.ylabel('Job Role', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Visualizing Distribution of Years at Company by Job Role

```python
# violin plot. It shows the distribution of "Years at Company" for each job role
plt.figure(figsize=(8, 6))
sns.violinplot(data=data, x='JobRole', y='YearsAtCompany', palette='muted', inner='box')
plt.title('Years at Company Distribution by Job Role', fontsize=16)
plt.xlabel('Job Role', fontsize=12)
plt.ylabel('Years at Company', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Illustrating Employee Attrition vs Overtime Hours

```python
# Stacked bar chart showing the percentage of employees experiencing attrition (yes or no) based on their overtime status
overtime_attrition = data.groupby(['OverTime', 'Attrition']).size().unstack()
overtime_attrition_percentage = (overtime_attrition.T / overtime_attrition.sum(axis=1)).T * 100
plt.figure(figsize=(10, 6))
overtime_attrition_percentage.plot(kind='bar', stacked=False, colormap='coolwarm', width=0.7, edgecolor='black')
plt.title('OverTime Impact on Attrition', fontsize=16)
plt.xlabel('OverTime', fontsize=12)
plt.ylabel('Percentage of Employees (%)', fontsize=12)
plt.legend(title='Attrition', fontsize=10)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Analysis of Job Role and Involvement Levels

```python
# Heatmap illustrates the relationship between job roles and job involvement levels
job_involvement_counts = data.groupby(['JobRole', 'JobInvolvement']).size().reset_index(name='Counts')
job_involvement_pivot = job_involvement_counts.pivot(index='JobRole', columns='JobInvolvement', values='Counts')
plt.figure(figsize=(8,5.5))
sns.heatmap(job_involvement_pivot, annot=True, fmt='d', cmap='Blues', linewidths=.5)
plt.title('Job Role and Job Involvement Levels')
plt.xlabel('Job Involvement Level')
plt.ylabel('Job Role')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

### Radar Chart for Department-wise Analysis of Satisfaction Level

```python
# Radar Chart of Average Satisfaction Levels Across Departments
satisfaction_columns = ['JobSatisfaction', 'EnvironmentSatisfaction', 'RelationshipSatisfaction']
avg_satisfaction = data2.groupby('Department')[satisfaction_columns].mean()
departments = avg_satisfaction.index
data = avg_satisfaction.values.T

# Plotting radar chart
fig, ax = plt.subplots(figsize=(8,5.5), subplot_kw=dict(polar=True))
angles = np.linspace(0, 2 * np.pi, len(departments), endpoint=False).tolist()
data = np.concatenate((data, data[:, [0]]), axis=1)
angles += angles[:1]
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles[:-1]), departments)
for i, col in enumerate(satisfaction_columns):
    ax.plot(angles, data[i], label=col)
    ax.fill(angles, data[i], alpha=0.25)
plt.title('Average Satisfaction Levels Across Departments')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```
